# Seleksi baris dengan banyak kriteria pada Pandas Data Frame

## Import modules

```python
import pandas as pd
```

## Persiapan Data Frame

```python
df = pd.read_csv('./data/titanicfull.csv')
df.head()
```

## Seleksi baris dengan banyak kriteria

```python
df[(df['sex']=='female') & (df['age']>=60) & (df['embarked']=='S') & (df['survived']==1)]
```

Kode di atas memiliki beberapa kriteria dalam menyeleksi data yang akan ditampilkan, yaitu:
* Jenis kelamin harus perempuan
* Umur mulai dari 60 tahun ke atas.
* Status `embarked` berupa `S`.
* Status `survived`-nya `1`.

Setiap kriteria harus diapit dengan tanda kurung.

```python
df[
    (df['sex']=='female') & 
    (df['age']>=60) & 
    (df['embarked']=='S') & 
    (df['survived']==1)
]
```

Kode di atas adalah contoh kode yang lebih mudah dibaca.

```python
kr1 = (df['sex']=='female')
kr2 = (df['age']>=60)  
kr3 = (df['embarked']=='S')
kr4 = (df['survived']==1)

df[kr1 & kr2 & kr3 & kr4]
```

Jika kriterianya dimasukkan ke dalam variable, maka saat penerapannya tidak perlu diapit tanda kurung.

# Parameter `header` dan `skiprows` pada method `read_csv`

## Import modules

```python
import pandas as pd
```

## Persiapan Data Frame

```python
df = pd.read_csv('./data/iris_error.csv')
df.head(8)
```

```python
df = pd.read_csv('./data/iris_error.csv', header=2, skiprows=[5, 6])
df.head(8)
```

Karena headernya ada di index ke 2, maka parameter `header` diberikan nilai `2`. Sedangkan parameter `skiprows` berfungsi untuk melewati baris-baris pada index tertentu, pada kasus ini terdapat pada index ke `5` dan `6`.

Karena parameter `header` sudah di set pada index ke 2, maka proses pembacaan data akan dimulai dari index **2** sehingga index ke **0** dan **1** tidak termasuk dalam data frame.

# Mengacak urutan baris pada Pandas Data Frame

## Import modules

```python
import pandas as pd
import numpy as np
```

## Persiapan Data Frame

```python
n_rows = 6
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)), columns = cols)

df
```

## Mengacak urutan baris pada Data Frame dengan method `sample`

```python
df.sample(frac=1.0, random_state=1)
```

Parameter `random_state` dispesifikasikan untuk menjamin hal yang dilakukan dapat dilakukan ulang (replicate). Sedangkan parameter `frac` diberi nilai 1.0 berarti akan dilakukan sampling dengan proporsi penuh.

```python
df.sample(frac=1.0, random_state=1).reset_index(drop=True) # Dilakukan untuk me-reset indexnya.
```
